## Scatter plot

```python
from bokeh.sampledata.iris import flowers
flowers.head(2)
```

```python
from bokeh.charts import Scatter
p = Scatter(flowers, x='petal_length', y='petal_width')
show(p)
```

```python
p = Scatter(flowers, x='petal_length', y='petal_width', color='species',
           legend='top_left')
show(p)
```

```python
p = Scatter(flowers, x='petal_length', y='petal_width', color='species',
           marker='species' , legend='top_left')
show(p)
```

## Bar chart

```python
from bokeh.sampledata.autompg import autompg
autompg.head(2)
```

```python
from bokeh.charts import Bar
p = Bar(autompg, label='cyl', values='mpg', agg='max',
       title='Max MPG by CYL', legend=None, tools='crosshair')
show(p)
```

```python
p = Bar(autompg, label='yr', values='mpg', agg='median', group='origin',
       title='Median MPG by YR grouped by ORIGIN', legend='top_left',
       tools='crosshair')
show(p)
```

```python
p = Bar(autompg, label='yr', values='mpg', agg='median', stack='origin',
       title='Median MPG by YR stacked by ORIGIN', legend='top_left',
       tools='crosshair')
show(p)
```

## Histograms

```python
import pandas as pd
import numpy as np

mu, sigma = 0, 0.5
normal = pd.DataFrame({'value': np.random.normal(mu, sigma, 1000),
                      'type': 'normal'})
lognormal = pd.DataFrame({'value': np.random.lognormal(mu, sigma, 1000),
                         'type': 'lognormal'})
df = pd.concat([normal, lognormal])
df[998:1002]
```

```python
from bokeh.charts import Histogram
hist = Histogram(df, values='value', bins=30)
show(hist)
```

```python
hist = Histogram(df[df['type']=='lognormal'], values='value',
                 bins=100, legend='top_left')
show(hist)
```

## Box Plot

```python
from bokeh.charts import BoxPlot
```

```python
p = BoxPlot(flowers, label='species', values='petal_width', tools='crosshair',
           color='#aa4444', xlabel='', ylabel='petal width, mm',
           title='Distribution of petals widths')
show(p)
```

```python
p = BoxPlot(flowers, label='species', values='petal_width', tools='crosshair',
           xlabel='', ylabel='petal width, mm',
           title='Distribution of petals widths',
           color='species')
show(p)
```
